In this section we are going to use Azure blob storage and Azure Data Lake Gen 2.


Azure Data Lake Storage Gen2 (ADLS Gen2) is a cloud-based repository for big data analytics built into Azure Blob storage. It allows users to interface with data using both file system and object storage paradigms. ADLS Gen2 is a core element of data analytics architectures and is designed to service multiple petabytes of information while sustaining hundreds of gigabits of throughput. <br>
ADLS Gen2 offers a variety of features, including: <br> 
File system semantics <br>
File-level security <br>
Scale <br>
Hierarchical storage <br>
Fine-grained security <br>
Compatibility with Hadoop <br>
<br>
ADLS Gen2 also offers low-cost, tiered storage, with high availability and disaster recovery capabilities. It has access tiers for storing data that's accessed frequently, infrequently, or rarely: <br>
Hot: Optimized for storing data that's accessed frequently <br>
Cool: Optimized for storing data that's infrequently accessed <br>
Cold: Optimized for storing data that is infrequently accessed or modified <br>
Archive: Optimized for storing data that's rarely accessed <br>
ADLS Gen2 also has a soft delete option that allows users to recover accidentally deleted files within a specified time period.

### What is Data lake ?
A data lake is a single, centralized repository where you can store all your data, both structured and unstructured. A data lake enables your organization to quickly and more easily store, access, and analyze a wide variety of data in a single location. With a data lake, you don't need to conform your data to fit an existing structure. Instead, you can store your data in its raw or native format, usually as files or as binary large objects (blobs).<br>

Azure Data Lake Storage is a cloud-based, enterprise data lake solution. It's engineered to store massive amounts of data in any format, and to facilitate big data analytical workloads. You use it to capture data of any type and ingestion speed in a single location for easy access and analysis using various frameworks. <br>
