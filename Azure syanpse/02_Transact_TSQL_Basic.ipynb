#### Examples of SQL Queries

1. GROUP BY :<br>
    SELECT ProductID, MAX(OrderQty) as "Max Order Qty", 
    SUM(OrderQty) as "Total Qty", 
    COUNT(OrderQty) as "Number of Qty" FROM SalesLT.SalesOrderDetail 
    GROUP BY ProductID;
<br> <br>
In group by clause all the columns which are not there in group by need to be use an aggregation command.
<br>In the SQL GROUP BY clause, we can use a column in the select statement if it is used in Group by clause as well. It does not allow any column in the select clause that is not part of GROUP BY clause.

2. PARTITION BY : <br>
SELECT ProductID, OrderQty,
SUM(OrderQty) OVER (PARTITION BY ProductID) as "Total Qty" 
FROM SalesLT.SalesOrderDetail 



3. Lead and LAG function
<br> Lead give the value of next row and lag gives value of previous row
<br> SQL Query : SELECT ProductID, OrderQty,
Lead(OrderQty) OVER (ORDER BY ProductID) as "Next Qty" ,
Lag(OrderQty) OVER (ORDER BY ProductID) as "Prev Qty"
FROM SalesLT.SalesOrderDetail 



4. WITH CLAUSE 
<br> This is like creating subset of query or like view but with no database obj
<br>SQL Query :
<br>WITH CTE_Products AS
(
SELECT OrderQty,ProductID,UnitPrice
FROM SalesLT.SalesOrderDetail
)<br>
SELECT * FROM CTE_Products;


5. Foreign Key Concept :
        
        CREATE TABLE Student
        (
            StudentID varchar(100) NOT NULL,
            StudentName varchar(1000),
            PRIMARY KEY (StudentID)
        ); 

        CREATE TABLE Course (
            CourseID varchar(100) NOT NULL,
            CourseName varchar(1000),
            Price real,    
            PRIMARY KEY (CourseID)
        );
        
        CREATE TABLE Orders (
            OrderID varchar(100) NOT NULL,
            CourseID varchar(100),
            StudentID varchar(100),    
            Discountpercent int,
            PRIMARY KEY (OrderID),
            FOREIGN KEY (StudentID) REFERENCES Student(StudentID),
            FOREIGN KEY (CourseID) REFERENCES Course(CourseID)
        );
